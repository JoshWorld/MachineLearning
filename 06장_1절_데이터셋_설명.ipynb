# 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
```

# 보스턴 데이터(datasets.load_boston)

```python
raw_boston = datasets.load_boston()
X_boston = pd.DataFrame(raw_boston.data)
y_boston = pd.DataFrame(raw_boston.target)
df_boston = pd.concat([X_boston,y_boston], axis=1)
```

```python
len(df_boston)
```

```python
df_boston.head()
```

```python
feature_boston = raw_boston.feature_names
print(feature_boston)
```

```python
col_boston=np.append(feature_boston, ['target'])
df_boston.columns = col_boston
df_boston.head()
```

```python
# 상관계수
corr_boston = df_boston.corr(method='pearson')
```

```python
corr_boston
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data = corr_boston, 
            annot=True, fmt = '.2f', 
            linewidths=.5, 
            cmap='Blues')
```

```python
plt.hist(df_boston['CRIM'])
plt.show()
```

```python
plt.title('CRIM histogram')
plt.hist(df_boston['CRIM'],bins=20)
plt.xlabel('CRIM')
plt.ylabel('Frequency')
plt.show()
```
